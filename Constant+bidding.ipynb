```python
clicks = 2034
imps = 2697738
cost = 216496239
costc = 216991

#Click-Through Rate
def CTR(clicks, imps):
   out = clicks/imps
   return out


#CVR(Conversion Rates)
def CVR(clicks, imps):
   out = cost/imps
   return out

#Average CPM(Cost Per Mille)
def CPM(cost, imps):
   out = (cost/imps)
   return out

#Average CPC(Cost Per Click)
def CPC(cost, clicks):
   cost = cost/1000 
   out = cost/clicks
   return out
```

# Constant bidding

```python
def constModel(constantbid):
    bidimpression = 0
    clicked = 0
    paycost = []
    cost = 0
    budget = 6250*1000
    
   
        
    for click, payprice in DFv[['click','payprice']].values:
            if constantbid > payprice:
                bidimpression += 1
                clicked += click
                cost += payprice
                paycost.append(payprice)
            if cost >= budget:
                break
    
    
    return bidimpression, clicked, cost
```

```python
nbids,numclick,paycost = constModel(273)
```

```python
print('Number of bids: %s' % nbids)
print('Number of clicks: %s' % numclick)
print('Total cost: %s ' % paycost)
```
