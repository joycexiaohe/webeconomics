# advertiser group

```python
dataset = pd.DataFrame()
```

```python
dataset['advertiser'] = np.sort(trainset.advertiser.unique())
dataset['impressions'] = trainset.groupby('advertiser').size().values
click = trainset.groupby('advertiser').click.value_counts()
dataset['clicks'] = click.iloc[click.index.get_level_values('click') == 1].values 
dataset['cost'] = trainset.groupby('advertiser').payprice.sum().values
```

```python
dataset['CTR'] = (dataset.clicks / dataset.impressions)
dataset['CPM'] = (dataset.cost / dataset.impressions)
dataset['eCPC'] = (dataset.cost / dataset.clicks)
```

# Data analysis on region 

```python
#impression and clicks based on each region
regionbased = pd.DataFrame()
regionbased = pd.DataFrame()
regionbased['region'] = np.sort(trainset.region.unique())

imps = trainset.groupby('region').advertiser.value_counts()

clicks = trainset.groupby('region').click.value_counts()
```

```python
clicks.iloc[clicks.index.get_level_values('click') == 0]=0

regionbased['impression'] = imps.groupby(level='region').sum().values
regionbased['clicks'] = clicks.groupby(level='region').sum().values
regionbased['CTR'] = (regionbased.clicks / regionbased.impression)
```
